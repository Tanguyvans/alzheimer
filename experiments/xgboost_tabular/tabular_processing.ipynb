```python
# Count NA values and total records for height and weight per patient
vitals_stats = vitals.groupby("PTID").agg({
    "VSHEIGHT": [lambda x: x.isna().sum(), "count"],
    "VSWEIGHT": [lambda x: x.isna().sum(), "count"]
})

# Calculate proportion of NA values
vitals_stats.columns = ["VSHEIGHT_na", "VSHEIGHT_total", "VSWEIGHT_na", "VSWEIGHT_total"]
vitals_stats["VSHEIGHT_na_prop"] = vitals_stats["VSHEIGHT_na"] / vitals_stats["VSHEIGHT_total"]
vitals_stats["VSWEIGHT_na_prop"] = vitals_stats["VSWEIGHT_na"] / vitals_stats["VSWEIGHT_total"]
```

We fill na values per patient , where we fill with mean for height and weight , and mode for units

```python
# Create a copy to avoid SettingWithCopyWarning
# Define columns to fill with their respective methods
numeric_cols = ['VSHEIGHT', 'VSWEIGHT']
categorical_cols = ['VSHTUNIT', 'VSWTUNIT']

# Fill numeric columns with mean per patient
for col in numeric_cols:
    vitals[col] = vitals.groupby('PTID')[col].transform(
        lambda x: x.fillna(x.mean())
    )

# Fill categorical columns with mode per patient
for col in categorical_cols:
    # Get mode for each patient
    modes = vitals.groupby('PTID')[col].agg(lambda x: x.mode()[0] if not x.mode().empty else None)
    
    # For each patient with NA values, fill with their mode
    for pid in vitals[vitals[col].isna()].PTID.unique():
        if pid in modes.index and not pd.isna(modes[pid]):
            vitals.loc[(vitals.PTID == pid) & (vitals[col].isna()), col] = modes[pid]

# Convert weight from pounds (VSWTUNIT=1) to kilograms (VSWTUNIT=2)
# 1 pound = 0.45359237 kilograms
pounds_mask = vitals.VSWTUNIT == 1

# Print stats before conversion
print(f"Before: pounds mean={vitals.loc[pounds_mask, 'VSWEIGHT'].mean():.2f}, kg mean={vitals[vitals.VSWTUNIT==2].VSWEIGHT.mean():.2f}")

# Vectorized operation to convert pounds to kg and update unit
vitals.loc[pounds_mask, 'VSWEIGHT'] *= 0.45359237
vitals.loc[pounds_mask, 'VSWTUNIT'] = 2

# Print stats after conversion
print(f"After: kg mean={vitals.VSWEIGHT.mean():.2f}, pounds count={sum(vitals.VSWTUNIT==1)}, kg count={sum(vitals.VSWTUNIT==2)}")




# Convert height from inches (VSHTUNIT=1) to centimeters (VSHTUNIT=2)
# 1 inch = 2.54 centimeters
inches_mask = vitals.VSHTUNIT == 1

# Print stats before conversion
print(f"Before: inches mean={vitals.loc[inches_mask, 'VSHEIGHT'].mean():.2f}, cm mean={vitals[vitals.VSHTUNIT==2].VSHEIGHT.mean():.2f}")

# Vectorized operation to convert inches to cm and update unit
vitals.loc[inches_mask, 'VSHEIGHT'] *= 2.54
vitals.loc[inches_mask, 'VSHTUNIT'] = 2


# Print stats after conversion
print(f"After: cm mean={vitals.VSHEIGHT.mean():.2f}, inches count={sum(vitals.VSHTUNIT==1)}, cm count={sum(vitals.VSHTUNIT==2)}")


vitals.VSHTUNIT.fillna(2,inplace=True)
vitals.VSWTUNIT.fillna(2,inplace=True)
vitals.VSHEIGHT.fillna(vitals.VSHEIGHT.mean(),inplace=True)
vitals.VSWEIGHT.fillna(vitals.VSWEIGHT.mean(),inplace=True)

```

# Option 1 : Merging dfs

```python

# Aggregate filtered_ptdemog
agg_ptdemog = ptdemog.groupby('PTID').agg({
    'PTRACCAT': lambda x: x.mode().iloc[0],  # categorical
    'PTGENDER': lambda x: x.mode().iloc[0],  # categorical
    'PTDOBYY': 'mean',  # numeric
    'PTHAND': lambda x: x.mode().iloc[0],  # categorical
    'PTMARRY': lambda x: x.mode().iloc[0],  # categorical
    'PTEDUCAT': 'mean',  # numeric
    'PTNOTRT': lambda x: x.mode().iloc[0],  # categorical
    'PTTLANG': lambda x: x.mode().iloc[0],  # categorical
}).reset_index()

# Aggregate med_history (all columns are categorical)
agg_med_history = med_history.groupby('PTID').agg({
    'MH14ALCH': lambda x: x.mode().iloc[0],
    'MH17MALI': lambda x: x.mode().iloc[0],
    'MH16SMOK': lambda x: x.mode().iloc[0],
    'MH15DRUG': lambda x: x.mode().iloc[0],
    'MH4CARD': lambda x: x.mode().iloc[0],
    'MHPSYCH': lambda x: x.mode().iloc[0],
    'MH2NEURL': lambda x: x.mode().iloc[0],
    'MH6HEPAT': lambda x: x.mode().iloc[0],
    'MH12RENA': lambda x: x.mode().iloc[0]
}).reset_index()

# Aggregate vitals
agg_vitals = vitals.groupby('PTID').agg({
    'VSWEIGHT': 'mean',
    'VSHEIGHT': 'mean',
    'VSWTUNIT': lambda x: x.mode().iloc[0],
    'VSHTUNIT': lambda x: x.mode().iloc[0]
}).reset_index()

# Aggregate mmse (cognitive test scores - using mean for cognitive scores)
agg_mmse = mmse.groupby('PTID').agg({
    'MMSCORE': 'mean'
}).reset_index()

# Aggregate neurobat (cognitive test scores - using mean for scores)
agg_neurobat = neurobat.groupby('PTID').agg({
    'TRAASCOR': 'mean',
    'TRABSCOR': 'mean',
    'TRABERRCOM': 'mean',
    'CATANIMSC': 'mean',
    'CLOCKSCOR': lambda x: x.mode().iloc[0],
    'BNTTOTAL': lambda x: x.mode().iloc[0],
    'DSPANFOR': lambda x: x.mode().iloc[0],
    'DSPANBAC': lambda x: x.mode().iloc[0]
}).reset_index()

# Aggregate cdr
agg_cdr = cdr.groupby('PTID').agg({
    'CDGLOBAL': lambda x: x.mode().iloc[0]
}).reset_index()

# Aggregate blchange
agg_blchange = blchange.groupby('PTID').agg({
    'BCFAQ': lambda x: x.mode().iloc[0],
    'BCDEPRES': lambda x: x.mode().iloc[0]
}).reset_index()

# Print the shapes of aggregated dataframes
print("\nShapes after aggregation:")
print(f"ptdemog: {agg_ptdemog.shape} (unique patients: {len(agg_ptdemog.PTID.unique())})")
print(f"med_history: {agg_med_history.shape} (unique patients: {len(agg_med_history.PTID.unique())})")
print(f"vitals: {agg_vitals.shape} (unique patients: {len(agg_vitals.PTID.unique())})")
print(f"mmse: {agg_mmse.shape} (unique patients: {len(agg_mmse.PTID.unique())})")
print(f"neurobat: {agg_neurobat.shape} (unique patients: {len(agg_neurobat.PTID.unique())})")
print(f"cdr: {agg_cdr.shape} (unique patients: {len(agg_cdr.PTID.unique())})")
print(f"blchange: {agg_blchange.shape} (unique patients: {len(agg_blchange.PTID.unique())})")
```

```python
adnidenoise_df.columns
```

```python
# Now we can merge all the aggregated dataframes

merged_df = agg_ptdemog.merge(agg_med_history, on='PTID', how='left')\
    .merge(agg_vitals, on='PTID', how='left')\
    .merge(agg_mmse, on='PTID', how='left')\
    .merge(agg_neurobat, on='PTID', how='left')\
    .merge(agg_cdr, on='PTID', how='left')\
    .merge(agg_blchange, on='PTID', how='left')\
    .merge(adnidenoise_df, right_on='subject_id',left_on='PTID', how='left')

print(f"\nFinal merged dataframe shape: {merged_df.shape}")
print(f"Number of unique patients in final dataframe: {len(merged_df.PTID.unique())}")

merged_df["BCDEPRES"].fillna(merged_df["BCDEPRES"].mode()[0],inplace=True)
merged_df["BCFAQ"].fillna(merged_df["BCFAQ"].mode()[0],inplace=True)


# Check for any missing values in the final merged dataframe
print("\nMissing values in merged dataframe:")
print(merged_df.isna().sum())
```
